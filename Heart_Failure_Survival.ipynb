#### Each square in the heatmap above shows the correlation between the variables on each axis. Correlation ranges from -1 to +1. Values closer to zero means there is no linear trend between the two variables. The closer the value is to 1 the columns are more positively correlated; i.e.  as one increases so does the other and the closer to 1 the stronger this relationship is. A correlation closer to -1 is similar, but instead of both increasing one variable will decrease as the other increases. The diagonals are all (1/dark blue) because those squares are correlating each variable to itself (so it's a perfect correlation). For the rest the larger the number and darker the color, the higher the correlation between the two variables. The plot is also symmetrical about the diagonal since the same two variables are being paired together in those squares.

### Feature selection with Hill climbing Technique for KNN

```python
x=heart_failure.drop('death_event',axis='columns')
#load the package
from sklearn.utils import shuffle
#checks the number of features
col_num=x.shape[1]
#store the selected features
new_ind=[]
#store the score for new state
cur_MaxScore=0.0
#initalize the oder of features to consider
col_Ind_Random=shuffle(range(0,col_num),random_state=1)
```

```python
for cur_f in range(0,col_num):
    #try each feature in the dataset
    new_ind.append(col_Ind_Random[cur_f])
    newdata=x.iloc[:,new_ind].copy()
    #train/test data
    x_train,x_test,Y_train,Y_test=train_test_split(newdata,target,test_size=0.20,random_state=3)
    #classifier name
    clf=KNeighborsClassifier(3)
    #fit the model in train data
    fit=clf.fit(x_train,Y_train)
    cur_Score=clf.score(x_test,Y_test)
    #if not better than the current state
    if cur_Score < cur_MaxScore :
        new_ind.remove(col_Ind_Random[cur_f])
    #otherwise this loop will run
    else :
        cur_MaxScore=cur_Score
        print("score with " + str(len(new_ind)) + " selected features : " + str(cur_Score))
        print("The feature indexes are " +str(new_ind))
```
